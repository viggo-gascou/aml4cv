```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

pretrained = pd.read_csv("../../results/loss/val_pretrained_cleaned.csv")
non_pretrained = pd.read_csv("../../results/loss/val_non_pretrained_cleaned.csv")
train_pretrained = pd.read_csv("../../results/loss/train_pretrained_cleaned.csv")
train_non_pretrained = pd.read_csv("../../results/loss/train_non_pretrained_cleaned.csv")



pretrained = pretrained[["epoch", "loss"]]
non_pretrained = non_pretrained[["epoch", "loss"]]
train_pretrained = train_pretrained[["epoch", "loss"]]
train_non_pretrained = train_non_pretrained[["epoch", "loss"]]
```

```python
plt.figure(figsize=(20 , 12))
# make linewidth thicker and dashes a bit further apart
sns.lineplot(data=non_pretrained, x="epoch", y="loss", label ="Baseline Val", color="mediumslateblue", linewidth=3)
sns.lineplot(data=train_non_pretrained, x="epoch", y="loss", label="Baseline Train", color="mediumslateblue", linestyle="--",linewidth=3)
sns.lineplot(data=pretrained, x="epoch", y="loss", label="Pretrained Val", color="orange", linewidth=3)
sns.lineplot(data=train_pretrained, x="epoch", y="loss", label="Pretrained Train", color="orange", linestyle="--", linewidth=3)
plt.title("Validation Loss Comparison: Pretrained vs Non-Pretrained Models", fontsize=30)
plt.tick_params(axis='both', labelsize=30)
plt.xlabel("epoch", fontsize=30)
plt.ylabel("Validation Loss", fontsize=30)
plt.ylim(0, None)
plt.legend(fontsize=20, loc='upper right')
plt.savefig("../../results/loss/loss_comparison.png")
plt.show()
```

```python


f1_non_pretrained = pd.read_csv("../../results/loss/non_pretrained_cleaned.csv")
f1_pretrained = pd.read_csv("../../results/loss/pretrained_cleaned.csv")


f1_pretrained
f1_non_pretrained
```

```python
# plot them on the same plot with seaborn
# set style to blue type style
sns.set_theme(style="white")
plt.figure(figsize=(20, 12))
sns.lineplot(data=f1_non_pretrained, x="epoch", y="f1", label="Baseline", color="mediumslateblue")
sns.lineplot(data=f1_pretrained, x="epoch", y="f1", label="Pretrained", color="orange")
plt.title("F1 Score Comparison", fontsize=30)
plt.tick_params(axis='both', labelsize=30)
plt.xlabel("epoch", fontsize=30)
plt.ylabel("F1 Score", fontsize=30)
plt.ylim(0, 1)
# increase legend font size
plt.legend(fontsize=20)
plt.savefig("../../results/f1_scores/f1_comparison.png")
plt.show()
```
